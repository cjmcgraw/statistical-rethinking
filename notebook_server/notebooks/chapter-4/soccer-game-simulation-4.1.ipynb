```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

plt.style.use('ggplot')
```

# Example prompt:

>You and a thousand of your closest friends line up on the halfway line of a soccer field. Each of you has a coin in your hand. At the sound of the whistle, you begin flipping the coins. Each time a coin comes up heads, that person moves one step towards the right-hand goal. Each time a coin comes up tails, that person moves one step towards the left-hand goal.
Each person flips the coin 16 times.

The rules appear to be the following:

1) binomial variable with p=0.5
2) on 0-0.49 -1
3) on 0.5-1.0 +1
4) everyone starts at 0.0

Onto the simulation code.

```python
def simulate_one_person(steps, p=0.5):
    """Simulates one persons path through the game.
    
    This could definitely be optimzied via vectorization,
    but it remains in its "simpler" form. I eat the cost
    of performance in favor of strong readability here
    """
    positions = np.zeros(steps + 1, np.int64)
    rolls = np.zeros(steps + 1, np.int64)
    current_position = 0
    for step in range(1, steps + 1):
        if np.random.binomial(1, p=p):
            rolls[step] = 1
        else:
            rolls[step] = -1
        positions[step] = np.sum(rolls)
    return rolls, positions
        

def simulate_steps(steps, number_of_people=10, p=0.5):
    """Simulate all steps of the game. returning a dataframe
    giving all data about each step of the game.
    
    There are fundamentally easier ways to simulate this
    utilizing an np array that is more heavily vectorized.
    I opted to use a for loop just for readability.
    """
    df = pd.DataFrame()
    for person in range(number_of_people):
        rolls, positions = simulate_one_person(steps, p=0.5)
        person_df = pd.DataFrame(dict(
            person=np.full(steps + 1, person),
            step=np.arange(0, steps + 1),
            roll=rolls,
            position=positions
        ))
        df = pd.concat([df, person_df])
    return df.set_index('person')
```

```python
df = simulate_steps(500, number_of_people=1000)
```

# Simulation complete

Now that we've completed the simulation lets plot the final positions as a graph of each person over time.

I have chosen to do 2500 people each doing 500 steps. The reason why is I wanted a sufficiently large sample to hide any rigid/lattice like strucuture. I wanted to see what happened as the sample size got large enough so that way patterns could be noticed or intuited.

```python
df.reset_index().set_index('step')
```

```python
plt.figure(figsize=(18, 8))
ax = plt.gca()
ax.set_xlabel('step')
ax.set_ylabel('distance from center in feet')
ax.set_title('simulation of the game by steps')
data = df.groupby('person').position.agg(list)
for x in data.sample(100):
    ax.plot(x, alpha=0.05, color='purple')

x, *_ = data.sample(1)

ax.plot(x, alpha=0.5, color='black')
```
